# I. Loading the dataset

```python
import numpy
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
import matplotlib
from matplotlib import pyplot
import numpy as np
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
#from catboost import CatBoostClassifier
import lightgbm as lgb
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
```

```python
# Load the Pima Indians dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
feature_names = names[:8]
data = pd.read_csv(url, names=names)
#Splitting the data into dependent and independent variables
```

```python
X = data.drop("class", axis=1)
y= data["class"]
# Splitting the data into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y,train_size=0.7)
```

# II. Training and evaluating the boosting ensembles

### 1) AdaBoost

```python
model_ada = AdaBoostClassifier()
model_ada.fit(X_train,y_train)
model_ada.score(X_test, y_test)
```

```python
model_ada = AdaBoostClassifier(n_estimators= 100)
model_ada.fit(X_train,y_train)
model_ada.score(X_test, y_test)
```

```python
param_grid = dict(n_estimators= range(50,200,10))
grid_search_ada = GridSearchCV(model_ada, param_grid, scoring="accuracy", n_jobs=-1, cv=10,
verbose=1)
grid_search_ada.fit(X, y)
means_ada = grid_search_ada.cv_results_['mean_test_score']
```

```python
X = data.drop("class", axis=1)
y= data["class"]
# Splitting the data into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y,train_size=0.7)
model_ada = AdaBoostClassifier(n_estimators= 100)
model_ada.fit(X_train,y_train)
model_ada.score(X_test, y_test)
param_grid = dict(n_estimators= range(50,200,10))
grid_search_ada = GridSearchCV(model_ada, param_grid, scoring="accuracy", n_jobs=-1, cv=10,
verbose=1)
grid_search_ada.fit(X, y)
means_ada = grid_search_ada.cv_results_['mean_test_score']
n_estimators=range(50,200,10)
def plot_results(model,means):
    scores = numpy.array(means).reshape(len(n_estimators))
    pyplot.plot(n_estimators, scores, label=''.join(str(model).split()[:2]) )
    pyplot.legend()
    pyplot.xlabel('n_estimators')
    pyplot.ylabel('accuracy')
    
plot_results(model_ada, means_ada)
```

### 2) Gradient Boosting

```python
model_gbm = GradientBoostingClassifier()
model_gbm.fit(X_train,y_train)
model_gbm.score(X_test, y_test)
```

```python
grid_search_gbm = GridSearchCV(model_gbm, param_grid, scoring="accuracy", n_jobs=-1, cv=10,
verbose=1)
grid_search_gbm.fit(X, y)
means_gbm = grid_search_gbm.cv_results_['mean_test_score']
plot_results(model_gbm,means_gbm )
```

### 3) XGBoost

```python
model_xgb = XGBClassifier()
model_xgb.fit(X_train,y_train)
y_pred = model_xgb.predict(X_test)
model_xgb.score(X_test, y_test)
```

```python
grid_search_xgb = GridSearchCV(model_xgb, param_grid, scoring="accuracy", n_jobs=-1, cv=10,
verbose=1)
grid_search_xgb.fit(X, y)
means_xgb = grid_search_xgb.cv_results_['mean_test_score']
plot_results(model_xgb,means_xgb)
```

# III. Comparing the different boosting methods

```python
models = ['AdaBoost','GBM','XGB']
means=numpy.concatenate((means_ada,means_gbm, means_xgb), axis=None)
scores = numpy.array(means).reshape(len(models), len(n_estimators))
for i, value in enumerate(models):
    pyplot.plot(n_estimators, scores[i], label=str(value))
pyplot.legend()
pyplot.xlabel('n_estimators')
pyplot.ylabel('accuracy')
pyplot.savefig('boosting_accuracy.png')
```

```python
models = ['AdaBoost','GBM','XGB']
means=numpy.concatenate((grid_search_ada.cv_results_['mean_score_time'],grid_search_gbm.cv_results_['mean_score_time'],grid_search_xgb.cv_results_['mean_score_time']), axis=None)
scores = numpy.array(means).reshape(len(models), len(n_estimators))
for i, value in enumerate(models):
    pyplot.plot(n_estimators, scores[i], label=str(value))
pyplot.legend()
pyplot.xlabel('n_estimators')
pyplot.ylabel('training_time')
pyplot.savefig('boosting_time.png')
```

```python
print("AdaBoost " +str(grid_search_ada.best_params_)+ "\t\t: accuracy = " + str(grid_search_ada.best_score_)+" / Training time = " + str(grid_search_ada.cv_results_['mean_score_time'][grid_search_ada.best_index_])+"(s)")
print("Grandient Boosting " +str(grid_search_gbm.best_params_)+ ": accuracy =" + str(grid_search_gbm.best_score_)+" / Training time = " + str(grid_search_gbm.cv_results_['mean_score_time'][grid_search_gbm.best_index_])+"(s)")
print("XGBoost "+str(grid_search_xgb.best_params_)+ "\t\t: accuracy = " + str(grid_search_xgb.best_score_)+" / Training time = " + str(grid_search_xgb.cv_results_['mean_score_time'][grid_search_xgb.best_index_])+"(s)")
```

```python
model_lgb = lgb.LGBMClassifier()
param_grid = {
    'n_estimators': range(50, 200, 10)
}
grid_search_lgb = GridSearchCV(model_lgb, param_grid, scoring="accuracy", n_jobs=-1, cv=10, verbose=1)
grid_search_lgb.fit(X, y)
means_lgb = grid_search_lgb.cv_results_['mean_test_score']
plot_results(model_lgb, means_lgb)
```

# V. Comparing the single classifiers and the boosting ensemble

```python
models = ['AdaBoost','GBM','XGB', 'LightGBM']
means=numpy.concatenate((means_ada, means_gbm, means_xgb, means_lgb), axis=None)
scores = numpy.array(means).reshape(len(models), len(n_estimators))
for i, value in enumerate(models):
    pyplot.plot(n_estimators, scores[i], label=str(value))
pyplot.legend()
pyplot.xlabel('n_estimators')
pyplot.ylabel('accuracy')
pyplot.savefig('boosting_accuracy.png')
```

```python
# Instanciation du modle
DT = DecisionTreeClassifier()

# Dfinition de la grille des hyperparamtres  tester
param_grid = {
    'max_depth': range(1, 21),  # Nombre maximum de niveaux de profondeur de l'arbre
}

# Recherche de grille d'hyperparamtres avec GridSearchCV
grid_search_DT = GridSearchCV(DT, param_grid, scoring="accuracy", n_jobs=-1, cv=10, verbose=1)
grid_search_DT.fit(X, y)

# Affichage des rsultats de la recherche de grille
means_DT = grid_search_DT.cv_results_['mean_test_score']
max_depths = param_grid['max_depth']
plt.plot(max_depths, means_DT, 'bo-')
plt.xlabel('max_depth')
plt.ylabel('Accuracy')
plt.title('Grid Search Results - Decision Tree')
plt.show()
```

```python
plotdata = pd.DataFrame({
"Ada Boost" : np.max(means_ada)*100,
"Gradient Boosting" : np.max(means_gbm)*100,
"XGBoost":np.max(means_xgb)*100,
"LightGBM":np.max(means_lgb)*100,
"Decision Tree":np.max(means_DT)*100}, index = ["Models"])
plotdata.plot(kind="bar",figsize=(8, 8))
plt.title("Comparaison of the Boosting ensembles and the single classifier")
plt.xlabel("Boosting ensemble and base learners model")
plt.ylabel("Accuracy (%)")
```

On remarque qu'en ne considerant que le meilleur de chaque model, le boosting n'amliore pas considerablement l'accuracy du base learner qui est le decision tree dans notre cas. On note egalement le fait que le gradient boosting est la meilleure configuration qui maximise l'accuracy.
